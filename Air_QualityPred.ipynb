```python
#loading the data from csv file to pandas dataframe

df= pd.read_csv('city_day.csv',parse_dates=["Date"])
df.sample(20)
```

```python
df.shape
```

```python
#create heatmap to check missing values
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='crest')
```

```python
df.info()
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Drop rows with missing values in the target variable (AQI_Bucket)
df = df.dropna(subset=['AQI_Bucket'])


# Check if there are still any missing values
print(df.isnull().sum())
```

```python
# Drop the 'Date' column
df = df.drop('Date', axis=1)
```

```python
df.head()
```

```python
# Identify numeric columns
numeric_cols = df.select_dtypes(include=[np.number]).columns

# Apply groupby and mean only to numeric columns
df[numeric_cols] = df.groupby("City")[numeric_cols].transform(lambda x: x.fillna(x.mean()))
```

```python
df.head()
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Identify numeric columns
numeric_cols = df.iloc[:, 2:14].select_dtypes(include=[np.number]).columns

# Fill missing values in numeric columns with the mean
df.loc[:, numeric_cols] = df.loc[:, numeric_cols].fillna(df[numeric_cols].mean())
```

```python
df.head(10)
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
df.shape
```

```python
# Drop the AQI column as it is not needed for classification
df = df.drop('AQI', axis=1)
```

```python
df.head(10)
```

```python
city_counts = df['City'].value_counts()

# Print the count of each city
print("City Counts:")
print(city_counts)
```

```python
from sklearn.preprocessing import LabelEncoder
# Create a copy of the DataFrame
df_encoded = df.copy()

# Apply label encoding to the 'City' column
label_encoder = LabelEncoder()
df_encoded['City'] = label_encoder.fit_transform(df_encoded['City'])
```

```python
df_encoded.head(5)
```

```python
df_encoded.tail(5)
```

```python
# Display the mapping between original city names and their corresponding encoded values
city_mapping = dict(zip(df['City'], df_encoded['City']))

# Print original city names and their corresponding encoded values
print("Mapping of Original City Names to Encoded Values:")
for city, encoded_value in city_mapping.items():
    print(f"{city}: {encoded_value}")
```

```python
# Assuming your DataFrame is named 'df'
unique_aqi_buckets = df_encoded['AQI_Bucket'].unique()

# Print the unique AQI bucket classes
print("Unique AQI Bucket Classes:")
for aqi_bucket in unique_aqi_buckets:
    print(aqi_bucket)
```

```python
from sklearn.preprocessing import LabelBinarizer

# Assuming your DataFrame is named 'df_encoded'
result_category = df_encoded['AQI_Bucket']

# Initialize LabelBinarizer
label_binarizer = LabelBinarizer()

# Apply one-hot encoding to the 'AQI_Bucket' column
result_one_hot = label_binarizer.fit_transform(result_category)

# Convert the result to a DataFrame
result_one_hot_df = pd.DataFrame(result_one_hot, columns=['AQI_Bucket_Binary'])

# Concatenate the binary-encoded column to the original DataFrame
df_encoded = pd.concat([df_encoded, result_one_hot_df], axis=1)

# Display the first 5 rows of the DataFrame with the binary-encoded column
print(df_encoded.head())
```
